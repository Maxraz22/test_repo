# Розв'язок

```python
import cv2
import numpy as np
```

### 1. Функція для зчитування зображення з диску та його відображення на екрані

```python
def show_image(file):
    # зчитати зображення з диску
    image = cv2.imread(file)  
    # перевірити, чи було успішно зчитано зображення
    if image is not None:
        cv2.imshow("Image", image)
        cv2.waitKey(0)
    else:
        print("Не вдалося зчитати зображення з диску")
```

```python
file = 'cat.jpg'
show_image(file)
```

### 2. Функція для збереження зображення на диск

```python
def save_image(image, filename):
    cv2.imwrite(filename, image)
    print("Зображення збережено у поточній директорії")
```

```python
image = cv2.imread("cat.jpg")
save_image(image, "image.jpg")
```

### 3. Функції для реалізації піксельних операцій

**3.1. Зміна контрастності (задання коефіцієнту контрастності)**

```python
import matplotlib.pyplot as plt
```

```python
def change_contrast(image, contrast):
    new_image = cv2.convertScaleAbs(image, alpha=contrast)
    return new_image
```

```python
image = cv2.imread("cat.jpg")
new_image = change_contrast(image, 0.4)
```

```python
plt.imshow(new_image)
plt.show()
```

**3.2. Зміна яскравості (задання коефіцієнту яскравості)**

```python
def change_brightness(image1, brightness):
    new1_image = cv2.convertScaleAbs(image, beta=brightness)
    return new1_image
```

```python
image1 = cv2.imread("cat.jpg")
new1_image = change_brightness(image1, 50)
```

```python
plt.imshow(new1_image)
plt.show()
```

**3.3. Зміна насиченості (задання коефіцієнту насиченості)**

```python
def change_saturation(image2, saturation):
    hsv_image = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)
    hsv_image[..., 1] = hsv_image[..., 1] * saturation
    new2_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
    return new2_image
```

```python
image2 = cv2.imread("cat.jpg")
new2_image = change_saturation(image2, 0.5)
```

```python
plt.imshow(new2_image)
plt.show()
```

**3.6. Поворот зображення на заданий кут (задання кута повороту в градусах)**

```python
def rotate_image(image5, angle):
    height, width = image5.shape[:2]
    center = (width/2, height/2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated_image = cv2.warpAffine(image5, rotation_matrix, (width, height))
    return rotated_image
```

```python
image5 = cv2.imread("cat.jpg")
rotated_image = rotate_image(image5, 45)
```

```python
plt.imshow(rotated_image)
plt.show()
```
