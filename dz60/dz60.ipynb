# Розв'язок

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Фільтр SOBEL 

```python
img = cv2.imread('lion.jpg')
```

```python
cv2.imshow('Original images', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('GRAY', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
```

```python
sobel_x = cv2.filter2D(gray, -1, kernel_x)
sobel_y = cv2.filter2D(gray, -1, kernel_y)
sobel = sobel_x + sobel_y
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.axis("off")
plt.imshow(img1)
print("Original images")
plt.show()
```

```python
sobel1 = cv2.cvtColor(sobel, cv2.COLOR_BGR2RGB)
```

```python
plt.axis("off")
plt.imshow(sobel1)
print("SOBEL")
plt.show()
```

### 2. Фільтр CANNY

```python
canny = cv2.Canny(gray, 50, 150)
```

```python
canny1 = cv2.cvtColor(canny, cv2.COLOR_BGR2RGB)
```

```python
plt.axis("off")
plt.imshow(img1)
print("Original images")
plt.show()
```

```python
plt.axis("off")
plt.imshow(canny1)
print("CANNY")
plt.show()
```

### 3. Фільтр ROBERTS

```python
kernel_x1 = np.array([[-1, 0], [0, 1]], dtype=np.float32)
kernel_y1 = np.array([[0, -1], [1, 0]], dtype=np.float32)
```

```python
rob_x = cv2.filter2D(gray, -1, kernel_x1)
rob_y = cv2.filter2D(gray, -1, kernel_y1)
rob = rob_x + rob_y
```

```python
plt.axis("off")
plt.imshow(img1)
print("Original images")
plt.show()
```

```python
rob1 = cv2.cvtColor(rob, cv2.COLOR_BGR2RGB)
```

```python
plt.axis("off")
plt.imshow(rob1)
print("ROBERTS")
plt.show()
```

### 4. Порівняння результатів

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].imshow(img1)
axs[0, 0].set_title('ORIGINAL')
axs[0, 1].imshow(sobel1)
axs[0, 1].set_title('SOBEL')
axs[1, 0].imshow(canny1)
axs[1, 0].set_title('CANNY')
axs[1, 1].imshow(rob1)
axs[1, 1].set_title('ROBERTS')

for ax in axs.flatten():
    ax.axis('off')

plt.show()
```

**За результатами застосування трьох різних фільтрів (SOBEL, CANNY і ROBERTS) у даному конкретному випадку для виділення границь на зображенні, візуально можемо зробити висновок, що найкращий результат забезпечив саме фільтр SOBEL. Основна різниця між застосованими фільтрами полягає у методах виявлення границь на зображенні та точності результуючих границь. SOBEL - це найпростіший з цих методів, ROBERTS - менш точний за SOBEL, а CANNY - найскладніший та найточніший метод з трьох. Таким чином, вибір фільтру для роботи із зображенням має враховувати переваги та недоліки цих методів обробки і найголовніше - особливості самого зображення та кінцевий результат, якого треба досягнути.**
