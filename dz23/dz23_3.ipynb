# Умова задачі

Два верстати-автомати виготовляють однакові вироби, які попадають на загальний конвеєр. Продуктивність першого автомата в два рази більше ніж продуктивність другого. Перший верстат-автомат виготовляє 60% виробів вищого гатунку, а другий - 84%.
а) Знайти ймовірність того, що навмання взятий з конвеєра виріб вищого гатунку.
б) Навмання взятий з конвеєра виріб має вищий гатунок. Знайти ймовірність того, що він виготовлений на першому верстаті.

# Розв'язок (a)

Подія А - навмання взятий з конвеєра виріб вищого гатунку. 

P(A1) - ймовірність того, що вироблений виріб є вищого гатунку (для першого верстата).

P(A2) - ймовірність того, що вироблений виріб є вищого гатунку (для другого верстата).

Враховуючи, що продуктивність першого верстата у два рази більша, ніж другого верстата, то ймовірність настання події A можна розрахувати наступним чином:

$ P(A) = \frac{2}{3} * P(A1) + \frac{1}{3} * P(A2) $

```python
p1 = 2/3 * 0.6 + 1/3 * 0.84
print("Ймовірність події, за якої навмання взятий з конвеєра виріб буде вищого гатунку:", round(p1, 2))
```

# Розв'язок (б)

Подія В - навмання взятий з конвеєра виріб буде вищого гатунку і вироблений на першому верстаті.

Враховуючи, що перший верстат удвічі продуктивніший, умовно вважатимемо, що за однаковий проміжлок часу на першому верстаті виготовляється 120 деталей (60 * 2). Тоді на другому виготовляється 84 деталі і 244 деталі виготовляється сумарно на двох верстатах (120 + 84). Таким чином, можна розрахувати ймовірність настання події B:

```python
p2 = (60 * 2) / (60 * 2 + 84)
print("Ймовірність події, за якої навмання взятий з конвеєра виріб буде вищого гатунку і вироблений на першому верстаті:", round(p2, 2))
```
