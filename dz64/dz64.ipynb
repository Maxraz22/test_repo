# Розв'язок

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im_src = cv2.imread("chess_2.png")
```

```python
pts_src = np.array([[168, 34], [529, 101], [409, 365], [189, 316]])
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show()
```

```python
im_dst = cv2.imread('chess_1.png')
```

```python
pts_dst = np.array([[465, 39], [463, 260], [202, 245], [192, 51]])
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.show()
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)
```

```python
h
```

```python
im_out = cv2.warpPerspective(im_src, h,
                             (im_dst.shape[1], 
                              im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```
