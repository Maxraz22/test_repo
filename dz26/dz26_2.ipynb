При вивченні випадкової величини Х у результаті 40 незалежних спостережень дістали вибірку: 10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7,9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12. Побудувати дискретний статистичний розподіл для цієї вибірки, а також полігон частот і F*(x). Знайти моду та медіану.

# Розв'язок #

### 1. Дискретний статистичний розподіл для вибірки ###

```python
import numpy as np
sample = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7,9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]
sample_set = set(sample)
sample_list = list(sample_set)
n_list = []
w_list = []
for i in sample_list:
    n = sample.count(i)
    n_list.append(n)
    w = round(n/len(sample), 3)
    w_list.append(w)

matrix = np.array([sample_list, n_list, w_list], dtype = float)
print(matrix)
```

Статистичний розподіл частот:

|$x_i$|6|7|8|9|10|11|12|13|14|
|-----|-|-|-|-|--|--|--|--|--|
|$n_i$|1|3|6|8|6 |6 |5 |3 |2 |

Статистичний розподіл відносних частот:

|$x_i$|6    |7    |8   |9  |10  |11  |12   |13   |14  |
|:----|:---:|:---:|:--:|:-:|:--:|:--:|:---:|:---:|:--:|
|$w_i$|0.025|0.075|0.15|0.2|0.15|0.15|0.125|0.075|0.05|

### 2. Полігон частот ###

```python
import matplotlib.pyplot as plt
plt.plot(sample_list, n_list)
plt.grid()
plt.xlabel(r"$x$")
plt.ylabel(r"$n$")
plt.show()
```

### 3. Емпірична функція розподілу $ F^*(x) $ ###

$ F^*(x) = \cfrac{n_x}{n} $,

де $ n_x = \displaystyle\sum_{x_i<x} n_i $

```python
plt.hist(sample, bins=len(sample), cumulative=True, histtype='bar')
plt.grid()
plt.xlabel(r'$x$')
plt.ylabel(r'$F^*(x)$')
plt.show()
```

### 4. Мода та медіана ###

Мода $ M_0 $ - це варіанта, яка має найбільшу частоту.

Отже, $ M_0 = 9 $

Медіана $ M_e $ - це варіанта, яка ділить ряд на дві частини, рівні за кількістю варіант.

При парному числі варіант $ k = 2*m $, медіана $ M_e = \cfrac{1}{2} * (x_m + x_{m+1}) $

У нашому випадку $ k = 40 $, отже $ M_e = \cfrac{1}{2} * (x_{20} + x_{21}) = 0,5 * (10 + 10) = 10 $

#### Визначення моди та медіани за допомогою спеціальних функцій Python:

```python
from statistics import mode, median
print("Мода статистичної вибірки:", mode(sample))
sample_for_median = sorted(sample)
print("Медіана статистичної вибірки:", median(sample_for_median))
```
